```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph([[1,2],[2,3],[3,4],[2,4],[1,3],[5,4],[5,1],[6,3],[4,6],[1,6]])
nx.draw(G,with_labels=True)
plt.show()
```

```python
A = nx.adjacency_matrix(G).todense()
```

```python
np.ones(A.shape[0])@A.T@A@np.ones(A.shape[1])
```

```python
np.trace(A@A@A)/6
```
